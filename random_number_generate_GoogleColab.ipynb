```
print ("i      "," Zi      ", " Ui      "," Zi*Zi      ")
for i in range(0,6):
  Zi= (random.randint(1111,9999))
 # print(Zi)
  Zi_square = int(Zi)**2
  x= str(Zi_square)

  if (len(x)!=8):
    add_zero = x.zfill(8)
    #print("add 0 : ",add_zero)
    y = add_zero[2:6]
    #print(y)
    Zi=y
    Ui=int(Zi)/10000
    #print(Ui)
    print(i,"    ",Zi,"    ",Ui,"    ",Zi_square)
  else:
    y= x[2:6]
    #print("without 0 : ",y)
    Zi=y
    Ui=int(Zi)/10000
    #print (Ui)
    print(i,"    ",Zi,"    ",Ui,"    ",Zi_square)
```
